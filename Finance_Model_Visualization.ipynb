```
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df= pd.read_excel('/content/drive/MyDrive/Mr Olowe/Data Extract for Machine Learning 20201213.xlsx')
```

```
df.head(5)
```

# The Target Value

```
df['status'].value_counts().plot.pie(autopct='%1.1f%%')
```

**Insight**: A larger number of the status value we are predicting appear to be in the RUNNING state, followed by SETTLED, with lesser percentage as PAST DUE

# Card Network

```
sns.countplot(x=df['card_network'])
```

**Insight:** Most clients are in possession of MasterCards

```
fig, axes = plt.subplots(1,2, figsize=(12,10))
cd= df.card_network.astype(str)
unique= np.unique(cd)

for i, ax in enumerate(axes.flatten()):
    values = df[df['card_network']==unique[i]]['status'].value_counts()
    values.plot.pie(ax=ax, autopct='%1.1f%%')
    ax.set_title(str(unique[i]))
    ax.set_ylabel('')

plt.show()
```

**Insight:** On comparing CARD NETWORK to the STATUS value, people with VISA card tends to have their loans settled than people with MASTERCards, ratio of each put into consideration
